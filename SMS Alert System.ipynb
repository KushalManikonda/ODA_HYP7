```python
from twilio.rest import Client

# Paste your credentials here
ACCOUNT_SID = "AC822e2b410e58ea3cc8631e91e50b0ae4"
AUTH_TOKEN = "0bf7f4b7c18144900363d1a8cf40393e"
TWILIO_PHONE = "+1(802) 810-8515"   # Twilio number
ADMIN_PHONE = "+919959583328"   # Your verified number
```

```python
SAFE_MODE = False   # Keep TRUE while testing
CONFIDENCE_THRESHOLD = 0.85


def calculate_severity(confidence, area):
    normalized_area = min(area / 10, 1)
    severity = (0.6 * confidence) + (0.4 * normalized_area)
    return severity


def send_alert(confidence, latitude, longitude, area):

    if confidence < CONFIDENCE_THRESHOLD:
        print("No alert triggered.")
        return

    severity = calculate_severity(confidence, area)

    message_body = f"""
OIL SPILL ALERT
Location: {latitude}, {longitude}
Area: {area} sq km
Confidence: {confidence*100:.2f}%
Severity Score: {severity:.2f}
- ODA Monitoring
"""

    if SAFE_MODE:
        print("SAFE MODE - SMS NOT SENT")
        print(message_body)
        return

    client = Client(ACCOUNT_SID, AUTH_TOKEN)

    message = client.messages.create(
        body=message_body,
        from_=TWILIO_PHONE,
        to=ADMIN_PHONE
    )

    print("SMS Sent Successfully")
    print("Message SID:", message.sid)
```
